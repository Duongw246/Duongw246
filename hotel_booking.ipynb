```python
import pandas as pd 
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

## LOADING DATA

```python
data = pd.read_csv('hotel_booking.csv')
data
```

```python
data.keys()
```

## DATA PREPROCESSING

```python
data.dtypes
```

```python
## checking the na list
data.isna().sum()
```

```python
## Reformat feature name and drop na values
data = data.rename(columns={'country code': 'country_code'})
data.country_code.dropna()
```

```python
data.agent.head(20)
```

#### DATA CLEANING

```python
## Removing irrelevant features
rev_irre_data = data.drop(['name', 'email','phone-number', 'credit_card', 'company', 'agent', 'reservation_status' , 'reservation_status_date'], axis=1)
```

#### DATA TRANSFORMATION

```python
encode = OrdinalEncoder()
for col in rev_irre_data.columns:
    if rev_irre_data[col].dtype == 'object':
        rev_irre_data[col] = encode.fit_transform(rev_irre_data[[col]])
rev_irre_data.dtypes
```

#### TRAIN TEST SPLIT

```python
X = rev_irre_data.drop(['is_canceled'], axis=1)
y = rev_irre_data['is_canceled']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

#### SCALING DATA

```python
scaler = MinMaxScaler()
scaled_train_data = scaler.fit_transform(X_train)
scaled_test_data = scaler.transform(X_test)
```

## TRAINING MODEL

```python
model = DecisionTreeClassifier(criterion="gini", max_depth=7, min_samples_split=9, random_state=10)
model.fit(scaled_train_data, y_train)
model.score(scaled_test_data, y_test)
```
