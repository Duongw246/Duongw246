```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
iris_dataset = load_iris()
print("Key of the dataset: \n{}".format(iris_dataset.keys()))
```

```python
print(iris_dataset['data'])
```

```python
print("Feature name: \n{}".format((iris_dataset['data'].shape)))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0
)
X_train
```

```python
df = pd.DataFrame(X_train, columns=iris_dataset['feature_names'])
df[:5]
```

```python
grr = pd.plotting.scatter_matrix(df, c=y_train, figsize=(15,15), marker='o', hist_kwds={'bins': 30, 'color':'red'},)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape: {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predict target name: {}".format(
    iris_dataset['target_names'][prediction]
))
```

```python
y_predict = knn.predict(X_test)
print("Test set prediction: {}".format(y_predict))
count = 0
print("Test set score: {:.2f}".format(np.mean(y_predict==y_test)))
```

```python
print('Test set score: {:.2f}'.format(knn.score(X_test, y_test)))
```
